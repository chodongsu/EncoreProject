```
content_data = pd.read_csv('daum_crawling_long.csv')
documents = [i for i in content_data['Content'].values]
```

```
N_data = ''
for i in documents:
    N_data += i
N_data
```

```
import nltk
nltk.download('punkt')

# 토큰화
from nltk.tokenize import word_tokenize
tag = word_tokenize(N_data)

# 토큰화
from nltk.tokenize import RegexpTokenizer
# [\w]+ 문자 또는 숫자가 1개 이상인 경우를 인식
retokenize = RegexpTokenizer("[\w]+")
retokenize.tokenize(N_data)
```
